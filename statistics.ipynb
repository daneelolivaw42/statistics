## Mean or Average
Mean or also average is the sum of all elements devided by the number of elements. Mean is the most popular measure of central tendency. It shows the average, but it does not show how far the spread is around this mean value or where in the dataset this mean value is located.  


**Pro: Generally the best measure of central tendency because, it utilizes all the scores.**    
**Con: Very sensitive to outliers (extreme scores).**    

$$ \bar{x} = \frac{1}{n} \left( \sum_{i=1}^{n}\right) $$


```python
data_table['mean'] = data_table.mean(axis=1)
data_table

plot_data = {
    'data': data_table,
    'x_data_column_name': data_table.index,
    'y_data_column_name': list(data_table.columns[0:]),
    'x_label': 'Time',
    'y_label': 'Value',
}

data_table
#p.scatter_plot(plot_data, style_plot)
```
